# Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In Pandas, a DataFrame is a two-dimensional, tabular data structure with labeled rows and columns, similar to a spreadsheet or SQL table, while a Series is a one-dimensional labeled array, like a single column in a DataFrame. 


```python
# Create Series
data = {"Name" : ["Alice", "Bob", "Claire"]}
series2 = pd.Series(data)
print(series2)
print(type(series2))

# Create DataFrame
df = pd.DataFrame({'Name': ["Alice", "Bob", "Claire"],
                   "Age" : [25,30,27],
                    'City': ['London', 'Paris', 'New York']})
print(df)
print(type(df))
```

from above example, series is a one-dimensional Series containing the names, and df is a two-dimensional DataFrame containing multiple columns with different data types (Name, Age, City)

# Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

All pandas data structures are value-mutable (the values they contain can be altered) but not always size-mutable.
1) Series : A one-dimensional labeled array capable of holding any data type. 
Value-mutable: You can change the values within a Series after it's created. 
Size-immutable: Once created, the length of a Series cannot be changed.
2) DataFrame : A two-dimensional labeled data structure with columns of potentially different types. 
Value-mutable: You can change the values within a DataFrame. 
Size-mutable: You can add or delete columns (and rows) after the DataFrame is created.
3) Panel : A Panel is a 3D array. It is not commonly used. It is not as easily displayed on screen or visualized as the other two because of its 3D nature. It is generally used for 3D time-series data.

# Q7. Create a DataFrame using multiple Series. Explain with an example

```python
courses = pd.Series(['Data Analysis', 'Data Sciesnce','Hadoop'], name = 'Courses')
fees = pd.Series([12000, 20000, 10000], name = 'Fees')
discount = pd.Series([2000, 5000, 1000], name = 'Discount')

df = pd.concat([courses, fees, discount], axis = 1)
df1 = pd.DataFrame([courses, fees, discount])
```
